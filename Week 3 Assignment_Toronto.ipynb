## Parse the Wiki Page using BEAUTIFULSOUP

```python
#Sitelink is the website we are asked to scrape.  We only need to pass the text to BeautifulSoup;
sitelink = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
webpage = requests.get(sitelink).text
```

```python
#Now we cook up some Beautiful Soup;
soup = BeautifulSoup(webpage)
```

```python
#The actual data we want is in an HTML Table object, so we parse it into a Pandas frame;
soup_table = str(soup.table)
```

```python
import pandas as pd
formatted_tables = pd.read_html(soup_table)
```

```python
#If we poke around a bit we see that the required table is in index 0 of formatted_tables;
df = formatted_tables[0]
```

```python
print(df.shape)
df.head()
```

## Data Cleansing

```python
#Now we begin our data cleansing steps;

#1. drop all rows where Borough = 'Not assigned';
df_1 = df[df.Borough != 'Not assigned']
```

```python
print(df_1.shape)
df_1.head(30)
```

```python
#2. combine rows with the same postal code into one row;  Turn off autosort;
df_2 = df_1.groupby(['Postal Code','Borough'], sort=False)
```

```python
#Then aggregate and separate by comma;
df_3=df_2.agg(', '.join)
```

```python
print(df_3.head())
df_3.shape
```

```python
#shape is messed up - reset index
df_3.reset_index(inplace=True)
df_3.shape
```

```python
df_3.head(30)
```

```python
#3. If neighborhood is not assigned, it's the same as the borough;
import numpy as np
df_3['Neighbourhood'] = np.where(df_3['Neighbourhood'] == 'Not assigned',df_3['Borough'], df_3['Neighbourhood'])
```

```python
#Part 1 is done - here's the entire dataset;
print(df_3.shape)
df_3.head(103)
```

```python
#Part 2 - now read in the table of latitudes and longitudes;
geos = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
geos.head()
```

```python
#Merge the geos into the master dataset on Postal Code;
df_4 = pd.merge(df_3,geos,on='Postal Code')
print(df_4.shape)
df_4.head()
```

```python
#Part 3;  filter on containing 'Toronto', cluster, and map;
df_5 = df_4[df_4['Borough'].str.contains('Toronto',regex=False)]
print(df_5.shape)
df_5.head(10)
```

## Map the Neighborhoods using FOLIUM

```python
#There are 39 areas to be clustered and mapped;
# create map of Toronto using latitude and longitude values
import folium
from geopy.geocoders import Nominatim
neighborhoods = df_5

address = 'Toronto, ON'

geolocator = Nominatim(user_agent="on_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
